```python
# Initialize H2O Environment

import h2o
from h2o.automl import H2OAutoML
h2o.init()
```

```python
# Load data into H2O

dataset_train = h2o.import_file("train.csv")
dataset_test = h2o.import_file("test.csv")
print (dataset_train,"\n\n")
print (dataset_test)
```

```python
# Use asfactor to understand we need output for classification problem
dataset_train['Survived'] = dataset_train['Survived'].asfactor()
```

```python
# Creating final  x  and  y  for prediction
y = "Survived"
x = dataset_train.columns
x.remove(y)
print (x)
```

```python
# Map data to x   and  y   with maximum model values
aml = H2OAutoML(max_models = 10, seed = 1)
aml.train(x = x, y = y, training_frame = dataset_train)
```

```python
# Getting all the models with accuracies (AUC)
lb = aml.leaderboard
print (lb)
```

```python
# Getting all information for the best model (accuracies,feature importance)
print (aml.leader)
```

```python
# Predict on test data
preds = aml.predict(dataset_test)
```

```python
# Generate final submission data
dataset_submission = h2o.import_file("submission.csv")
frame = preds['predict']
dataset_submission['Survived']=frame
print(dataset_submission)
```

```python
# Generate final submission file 
h2o.download_csv(dataset_submission,"h2o_submission.csv")
```

```python
# Generate final model which we can use for the deployment part
h2o.save_model(aml.leader, path = "h2o_first_prediction")
```
